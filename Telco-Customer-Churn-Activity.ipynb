# 3) Compare 'tenure' with 'Total Charges'

```python
plt.scatter(data['tenure'],data['TotalCharges'])
```

We have some whitespaces in  'TotalCharges' or 'tenure'
So, We have to find it

```python
df=data
```

```python
df.info()
```

```python
df.loc[df['tenure']==' ']
```

'tenure' Has no whitespaces

```python
df.loc[df['TotalCharges']==' ']
```

'TotalCharges' Has contain 11 Whitespaces

```python
#Take the indices having Whitespaces in 'TotalCharges'
indices=df.loc[df['TotalCharges']==' '].index
```

```python
#Drop the Rows having Whitespaces in 'TotalCharges'
df.drop(indices, inplace = True)
```

```python
df.loc[df['TotalCharges']==' ']
```

Row Containing Whitespaces were droped

```python
df
```

```python
#Comparison Plot: tenure with Total Charges
plt.figure(figsize=(8,8))
plt.scatter(df['tenure'],df['TotalCharges'],s=10)
plt.title('Comparison Plot: tenure with Total Charges',fontweight='bold',fontsize=15)
plt.xlabel('Tenure')
plt.ylabel('TotalCharges')
```

# 4) Findout which contract preffered by the senior citizen.

```python
#Picking out Senior Citizen's Data
sCitizen=data[data['SeniorCitizen']==1]
sCitizen
```

```python
#Counts of varios contract preffered by the senior citizen
sCitizen['Contract'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.title('Contract Preffered by the Senior Citizen',fontweight='bold',fontsize=15)
plt.pie(sCitizen['Contract'].value_counts(),labels=['Month-to-month','One year','Two year'],shadow=True,explode=(0.03, 0.03, 0.03),autopct = '%1.f%%')
plt.legend(title='Contracts')
plt.show()
```

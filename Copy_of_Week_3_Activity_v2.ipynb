# **4.5**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
plt.hist(df2['Age'])
plt.title('Distribution of Age',fontsize=20,color='blue')
plt.xlabel('Age',color='green')
plt.ylabel('Age Count',color='green')
plt.yticks(np.arange(0,260,10),color='red')
plt.xticks(np.arange(20,82,2),color='red')
plt.grid()
```

# **4.6**

```python
plt.figure(figsize=(15,7))
plt.scatter(df2['Glucose'],df2['BMI'],s=10)
plt.title('BMI VS GLUCOSE',fontsize=20,color='blue')
plt.xlabel('BMI',color='green')
plt.xticks(np.arange(0,210,10),color='red')
plt.ylabel('Glucose',color='green')
plt.yticks(color='red')
```

# **4.7**

```python
df2.replace(min(df2['BMI']),df2['BMI'].mean())
```

# **4.8**

```python
import seaborn as sns
plt.figure(figsize=(8,8))
sns.boxplot(x='Outcome', y='Glucose',data=df2)
plt.title('Distributions of Glucose for each outcome.',fontsize=20,color='blue')
plt.yticks(color='red')
plt.xticks(color='Green')
```

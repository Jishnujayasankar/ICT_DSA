# 3. Who scores the most on average for math, reading and writing based on
# ● Gender
# ● Test preparation course


```python
#Creating a new Column Contain Average Mark Of Each Student
data['Average Score']=((data['math score']+data['reading score']+data['writing score'])/3)
#Finding the Maximum Average Mark In the table
print("Maximum Average=",data['Average Score'].max())
```

```python
#Creating A subset dataframe from main dataframe,Which Contains only students who got maximum Average
maxAvg=data['Average Score']==data['Average Score'].max()
avgTable=data[maxAvg]
#Finding male student who got maximum Average
print('Male Student Who Got Maximum Average Score')
male=avgTable['gender']=='male'
avgTable[male]
```

```python
#Finding Female student who got maximum Average
print('Female Student Who Got Maximum Average Score')
female=avgTable['gender']=='female'
avgTable[female]
```

```python
#Finding Test Preperation Course Completed Student Who Got Maximum Average Score 
print('Test Preperation Course Completed Student Who Got Maximum Average Score')
completed=avgTable['test preparation course']=='completed'
avgTable[completed]
```

```python
#Finding Test Preperation Course Not Completed Student Who Got Maximum Average Scor
print('Test Preperation Course Not Completed Student Who Got Maximum Average Score')
ncompleted=avgTable['test preparation course']=='none'
avgTable[ncompleted]
```

# 4. What do you think about the scoring variation for math, reading and writing based on
#  ● Gender
#  ● Test preparation course

```python
#Filter Table Of Male Students
mDf=data[data['gender']=='male']
#Filter Table Of Female Students
fDf=data[data['gender']=='female']
```

```python
plt.figure(figsize=(15,8))

#Score Variation Plots Of Male Sudents
plt.subplot(2,3,1)
mDf['math score'].hist()
plt.title('Mathscore Variation',color='green')
plt.ylabel('Male Students',color='red')

plt.subplot(2,3,2)
mDf['reading score'].hist()
plt.title('Reading Score',color='green')

plt.subplot(2,3,3)
mDf['writing score'].hist()
plt.title('Writing Score',color='green')

#Score Variation Plots Of Female Students
plt.subplot(2,3,4)
fDf['math score'].hist()
plt.ylabel('Female Students',color='red')

plt.subplot(2,3,5)
fDf['reading score'].hist()

plt.subplot(2,3,6)
fDf['writing score'].hist()

plt.suptitle('Scoring Varition-Gender Base',fontsize=20,fontweight='bold')
plt.show()
```

```python
#Filter Table Of Preperation Course Completed Students
cDf=data[data['test preparation course']=='completed']
#Filter Table Of Preperation Course Not Completed Students
nDf=data[data['test preparation course']=='none']
```

```python
plt.figure(figsize=(15,8))

#Score Variation Plots Of Preperation Course Completed Students
plt.subplot(2,3,1)
cDf['math score'].hist()
plt.title('Mathscore Variation',color='green')
plt.ylabel('Preperation Course Completed Students',color='red')

plt.subplot(2,3,2)
cDf['reading score'].hist()
plt.title('Reading Score',color='green')

plt.subplot(2,3,3)
cDf['writing score'].hist()
plt.title('Writing Score',color='green')

#Score Variation Plots Of Preperation Course Not Completed Students
plt.subplot(2,3,4)
nDf['math score'].hist()
plt.ylabel('Preperation Course Not Completed Students',color='red')

plt.subplot(2,3,5)
nDf['reading score'].hist()

plt.subplot(2,3,6)
nDf['writing score'].hist()

plt.suptitle('Scoring Varition-Test Preperation Course Base',fontsize=20,fontweight='bold')
plt.show()
```

# 5.The management needs your help to give bonus points to the top 25% of students based on their maths score, so how will you help the management to achieve this.

```python
df=data
#Sort the table by column Maths Score in Descending Order
df=df.sort_values(by='math score',ascending=False)
df
```

```python
#Total row count
df.shape[0]
```

```python
#25% of 1000 is
(25*df.shape[0])/100
```

So in the sorted dataframe we can select first 250 students for the bonus mark

```python
#Creating A datset which contain 25% student who have top math score
df=df.head(250)
df
```

These students are eligible for the bonus mark

Consider The Bonus Point Are 
    90< =10,
    80< =8,
    70< =6,

```python
#Create A New Column Contains Bonus Mark for students based on their mark score
condition=[df['math score']>=90,df['math score']>=80,df['math score']>=70]
choice=[10,8,6]
df['Bonus Mark']=np.select(condition,choice)
```

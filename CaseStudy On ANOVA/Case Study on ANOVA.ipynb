# ● Check whether all the zones generate the same amount of sales.

Null Hypothesis  
H0: mean(Zone - A) = mean(Zone - B) = mean(Zone - C) = mean(Zone - D)

```python
#Prepare a ANOVA Test
from scipy import stats
f_value, p_value = stats.f_oneway(df['Zone - A'],df['Zone - B'],df['Zone - C'],df['Zone - D'])
```

```python
#F value and P value
print('F Value =',f_value.round(7))
print('P Value =',p_value.round(7))
```

P value is less than 0.05. ie, We force to reject the H0.  
So we can conclude that, each zone generates different amount of sales
